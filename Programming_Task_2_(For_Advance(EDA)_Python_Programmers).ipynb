```python
#  Defanging an IP Address
def defangIPaddr(address: str) -> str:
    return address.replace('.', '[.]')

# Example usage:
input_ip = "255.100.50.0"
output_ip = defangIPaddr(input_ip)
print(output_ip)
```

```python
# Find Numbers with Even Number of Digits
def findNumbers(nums):
    count = 0
    for num in nums:
        if len(str(num)) % 2 == 0:
            count += 1
    return count

# Example 1
nums1 = [12, 345, 2, 6, 7896]
print(findNumbers(nums1))  # Output: 2

# Example 2
nums2 = [555, 901, 482, 1771]
print(findNumbers(nums2))  # Output: 1

# Convert each number to a string to easily count its digits using len().
# Check if the number of digits is even using % 2 == 0.
# Count how many numbers satisfy this condition.
```

```python
# Number of Good Pairs
def numIdenticalPairs(nums):
    count = 0
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] == nums[j]:
                count += 1
    return count

# Example 1
print(numIdenticalPairs([1, 2, 3, 1, 1, 3]))  # Output: 4

# Example 2
print(numIdenticalPairs([1, 1, 1, 1]))        # Output: 6

# Example 3
print(numIdenticalPairs([1, 2, 3]))           # Output: 0

# Explanation:
# A good pair is defined as a pair (i, j) where i < j and nums[i] == nums[j].
# The brute-force approach checks all valid (i, j) pairs — acceptable here because n ≤ 100.
# Time Complexity: O(n²)
# Space Complexity: O(1)
```

```python
# How Many Numbers Are Smaller Than the Current Number
def smallerNumbersThanCurrent(nums):
    # Count frequency of each number (since nums[i] is between 0 and 100)
    count = [0] * 101
    for num in nums:
        count[num] += 1

    # Compute prefix sum: smaller[i] = total count of numbers < i
    smaller = [0] * 101
    for i in range(1, 101):
        smaller[i] = smaller[i - 1] + count[i - 1]

    # Build result
    return [smaller[num] for num in nums]

# Example 1
print(smallerNumbersThanCurrent([8, 1, 2, 2, 3]))  # Output: [4, 0, 1, 1, 3]

# Example 2
print(smallerNumbersThanCurrent([6, 5, 4, 8]))     # Output: [2, 1, 0, 3]

# Example 3
print(smallerNumbersThanCurrent([7, 7, 7, 7]))     # Output: [0, 0, 0, 0]

# How It Works Explanation:
# Since the values are constrained to 0 <= nums[i] <= 100, we use counting sort ideas.
# count[num] = how many times num appears.
# smaller[num] = total numbers less than num = sum of counts from 0 to num-1.
# Then, for each num in the original array, just look up smaller[num].
# Time Complexity: O(n + 100) → O(n)
# Space Complexity: O(100) → O(1) (since 101 is constant)

# This is much more efficient than the naive O(n²) double loop!
```

```python
# Subtract the Product and Sum of Digits of an Integer
def subtractProductAndSum(n: int) -> int:
    product = 1
    total = 0
    while n > 0:
        digit = n % 10
        product *= digit
        total += digit
        n //= 10
    return product - total

# Example 1
print(subtractProductAndSum(234))   # Output: 15

# Example 2
print(subtractProductAndSum(4421))  # Output: 21
# How It Works:
# Extract each digit using n % 10.
# Update product (starts at 1) and sum (starts at 0).
# Remove the last digit with n //= 10.
# Return product - sum.
# Time Complexity: O(d), where d is the number of digits (max 6 for n ≤ 10^5)
# Space Complexity: O(1)
```

```python
# XOR Operation in an Array
def xorOperation(n: int, start: int) -> int:
    result = 0
    for i in range(n):
        result ^= (start + 2 * i)
    return result

# Example 1
print(xorOperation(5, 0))   # Output: 8

# Example 2
print(xorOperation(4, 3))   # Output: 8
# How It Works:
# Build the array virtually using the formula: nums[i] = start + 2 * i
# Use XOR (^=) to accumulate the result
# No need to store the full array — just compute on the fly
# Time Complexity: O(n)
# Space Complexity: O(1)

# This is efficient and meets all constraints (n ≤ 1000).
```
